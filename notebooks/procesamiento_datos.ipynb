```python
import pandas as pd
import ast
import numpy as np
```

```python
df_ = pd.read_csv('../datos/preprocesados/df.csv')
df_matriz = pd.read_csv('../datos/preprocesados/df_matriz.csv')

df_.head(50)
```

```python
def safe_eval(item):
    try:
        return ast.literal_eval(item)
    except (ValueError, SyntaxError) as e:
        print(f"Error al evaluar: {e} - Item: {item}")
        return None 
```

```python

def convert_to_float32(matrix):
    if matrix is None:
        return None
    return np.array(matrix, dtype=np.float32)
```

```python
df_matriz['matriz'] = df_matriz['matriz'].apply(lambda x: convert_to_float32(safe_eval(x)))
```

```python
print(df_matriz.dtypes)  # Para verificar los tipos de datos de las columnas

# Verificar los primeros elementos de la columna 'matriz'
print(df_matriz['matriz'].head())

# Verificar el contenido de una celda espec√≠fica
print(df_matriz['matriz'].iloc[0]) 
```

```python
for matriz in df_matriz['matriz']:
    print(type(matriz), matriz.dtype) 
```

```python
df_final = pd.merge(df_, df_matriz, on='imagen', how='inner')
df_final.head()
# Guardar dataframe final
df_final.to_csv('../datos/df_final.csv', index=False)
```

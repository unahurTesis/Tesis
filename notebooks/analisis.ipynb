```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_pickle('../datos/df_final_limpio.pkl')
df
```

```python
print(df.info())
print("\nPrimeras filas del DataFrame:")
print(df.head())
```

```python
print("\nValores faltantes por columna:")
print(df.isnull().sum())
```

```python
print("\nEstadísticas descriptivas de la variable objetivo:")
print(df['matriz'].describe())

plt.figure(figsize=(10, 6))
df['matriz'].hist()
plt.title('Distribución de la variable objetivo')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

```python
numeric_columns = df.select_dtypes(include=[np.number]).columns
for col in numeric_columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot de {col}')
    plt.show()
```

```python
# Verificación de la cantidad de datos
print("\nDimensiones del DataFrame:")
print(df.shape)
```

```python
# Detección de valores atípicos (ejemplo para la variable objetivo)
from scipy import stats

z_scores = np.abs(stats.zscore(df['matriz']))
print("\nNúmero de valores atípicos en la variable objetivo (|z-score| > 3):")
print(np.sum(z_scores > 3))
```
